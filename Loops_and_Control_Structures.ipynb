## Section 1: Control Structures (If-Else)

```

# Grade Calculator
marks = 85
if 0 <= marks <= 100:
    if marks >= 90:
        print("Grade A")
    elif marks >= 80:
        print("Grade B")
    elif marks >= 70:
        print("Grade C")
    elif marks >= 60:
        print("Grade D")
    else:
        print("Grade F")
else:
    print("Invalid marks")

# Leap Year Checker
year = 2024
if year % 400 == 0:
    print("Leap Year")
elif year % 100 == 0:
    print("Not Leap Year")
elif year % 4 == 0:
    print("Leap Year")
else:
    print("Not Leap Year")

# Triangle Type
a, b, c = 5, 5, 5
if a+b>c and b+c>a and a+c>b:
    if a==b==c:
        print("Equilateral")
    elif a==b or b==c or a==c:
        print("Isosceles")
    else:
        print("Scalene")
else:
    print("Invalid Triangle")

# BMI Calculator
weight, height = 70, 1.75
bmi = weight / (height**2)
if bmi < 18.5:
    print("Underweight")
elif bmi < 25:
    print("Normal")
elif bmi < 30:
    print("Overweight")
else:
    print("Obese")
```

## Section 2: For Loops

```

# Multiplication Table
n = 5
for i in range(1,11):
    print(n, "x", i, "=", n*i)

# Factorial
num = 5
fact = 1
for i in range(1, num+1):
    fact *= i
print("Factorial:", fact)

# Sum of Evens
n = 10
s = 0
for i in range(2, n+1, 2):
    s += i
print("Sum of evens:", s)

# Fibonacci
n = 7
a, b = 0, 1
for _ in range(n):
    print(a, end=" ")
    a, b = b, a+b
print()

# Prime Checker
n = 29
is_prime = n > 1
for i in range(2, int(n**0.5)+1):
    if n % i == 0:
        is_prime = False
        break
print("Prime:", is_prime)
```

## Section 3: While Loops

```

# Reverse Number
n = 12345
rev = 0
temp = n
while temp > 0:
    rev = rev*10 + temp%10
    temp //= 10
print("Reverse:", rev)

# Sum of Digits
n = 12345
s = 0
while n > 0:
    s += n % 10
    n //= 10
print("Sum of digits:", s)

# Palindrome
num = 121
temp = num
rev = 0
while temp > 0:
    rev = rev*10 + temp%10
    temp //= 10
print("Palindrome:", num == rev)
```

## Section 4: Nested Loops (Patterns)

```

n = 5
for i in range(1, n+1):
    print("*" * i)

for i in range(n, 0, -1):
    print("*" * i)

for i in range(1, n+1):
    print(" "*(n-i) + "*"*(2*i-1))

for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end="")
    print()
```

## Section 5: Break & Continue

```

# First multiple of 7 and 13
for i in range(1,101):
    if i%7==0 and i%13==0:
        print(i)
        break

# Skip multiples of 3
for i in range(1,51):
    if i%3==0:
        continue
    print(i, end=" ")
print()
```

## Section 6â€“11: Advanced Problems (Representative Solutions)

```

# GCD
a, b = 48, 18
while b:
    a, b = b, a%b
print("GCD:", a)

# LCM
x, y = 12, 18
a, b = x, y
while b:
    a, b = b, a%b
print("LCM:", (x*y)//a)

# Perfect Number
n = 28
s = 0
for i in range(1, n):
    if n%i==0:
        s+=i
print("Perfect:", s==n)

# Pascal Triangle
rows = 5
for i in range(rows):
    num = 1
    print(" "*(rows-i), end="")
    for j in range(i+1):
        print(num, end=" ")
        num = num*(i-j)//(j+1)
    print()

# Caesar Cipher (Encrypt)
text = "Hello"
shift = 3
result = ""
for ch in text:
    if ch.isalpha():
        base = ord('A') if ch.isupper() else ord('a')
        result += chr((ord(ch)-base+shift)%26 + base)
    else:
        result += ch
print("Encrypted:", result)
```
