## String Assignment (Python)

Run each cell step by step.

```python

s = input("Enter a string: ")

print("\n1. Length of the string:", len(s))
print("2. Uppercase:", s.upper())
print("   Lowercase:", s.lower())

vowels = "aeiouAEIOU"
vowel_count = 0

for letter in s:
    if letter in vowels:
        vowel_count += 1

print("3. Number of vowels:", vowel_count)

reverse = ""

for letter in s:
    reverse = letter + reverse

if s == reverse:
    print("4. The string is a Palindrome")
else:
    print("4. The string is NOT a Palindrome")

print("5. Number of words:", len(s.split()))
print("6. Replace spaces with _ :", s.replace(" ", "_"))
print("7. Reversed string:", reverse)
```

```python

s2 = "   Learning Python is simple and Python is powerful   "

print("\n--- Using Given String ---")

print("1. Remove spaces:", s2.strip())

ch = input("Enter starting character: ")
print("2. Starts with given character:", s2.strip().startswith(ch))

ch2 = input("Enter character to find index: ")
print("3. Index of character:", s2.find(ch2))

ch3 = input("Enter character to count: ")
print("4. Count of character:", s2.count(ch3))

print("5. Split into words:", s2.strip().split())

words = ["Learning", "Python", "is", "easy"]
print("6. Join words:", " ".join(words))

print("7. Replace word:", s2.replace("Python", "Java"))
```
